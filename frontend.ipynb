```python
import json
from pathlib import Path
```

```python
PDF_CATEGORY_MAP = {
    "01_Radha_Krishna_Art_Poster": {"category": "Radha Krishna"},
    "02_Radha_Krishna_Art_Poster_Vertical": {"category": "Radha Krishna"},
    "02_Ram_Catalogue_PDF_Brochure": {"category": "Ram"},
    "03_Mixed_Art_Poster_Chinese_Scenery": {"category": "Chinese Scenery"},
    "04_Ganesh_Art_Poster": {"category": "Ganesh"},
    "08_Ambedkar_Buddha_Art_Poster": {"category": "Ambedkar Buddha"},
    "08_Horse_Art_Poster": {"category": "Horse"},
    "09_Ambedkar_Art_Poster": {"category": "Ambedkar"},
    "29_Chinese_Scenery_Modern_Art_Poster": {"category": "Chinese Scenery"},
    "30_Chinese_Scenery_Modern_Art_Poster": {"category": "Chinese Scenery"},
    "31_Chinese_Scenery_Modern_Art_Poster": {"category": "Chinese Scenery"},
    "38_Nature_Art_Poster": {"category": "Nature"},
    "39_Modern_Art_Poster_2023": {"category": "Modern"},
    "40_Mahakal_Art_Poster_2023": {"category": "Mahakal"},
    "43_Modern_Art_Poster": {"category": "Modern"},
    "44_Modern_Art_Poster": {"category": "Modern"},
    "45_Tirupati_Balaji_Art_Poster": {"category": "Tirupati Balaji"},
    "46_Ram_Darbar_Art_Poster": {"category": "Ram Darbar"},
    "47_Shiva_Family_Art_Poster": {"category": "Shiva Family"},
    "48_Hanuman_Art_Poster": {"category": "Hanuman"},
    "49_LGS_Art_Poster_New": {"category": "Lakshmi Ganesh Saraswati"},
    "51_Buddha_Art_Poster": {"category": "Buddha"},
    "61_Mixed_Gods_Art_Poster": {"category": "Mixed Gods"},
    "72_Buddha_Art_Poster": {"category": "Buddha"},
    "75_Khatu_Shyam_Art_Poster": {"category": "Khatu Shyam"},
}
```

```python
def generate_pdfs_metadata(
    posters_json_dir="frontend/json",
    output_path="frontend/json/pdfs_metadata.json",
):
  posters_json_dir = Path(posters_json_dir)
  output_path = Path(output_path)

  result = {}

  for json_file in posters_json_dir.glob("*.json"):
    pdf_name = json_file.stem

    if pdf_name not in PDF_CATEGORY_MAP:
      raise KeyError(f"Missing category for PDF: {pdf_name}")

    with open(json_file, "r", encoding="utf-8") as f:
      data = json.load(f)

    result[pdf_name] = {
        "category": PDF_CATEGORY_MAP[pdf_name]["category"],
        "total_posters": len(data["posters"]),
    }

  with open(output_path, "w", encoding="utf-8") as f:
    json.dump(result, f, indent=2, ensure_ascii=False)
```

```python
generate_pdfs_metadata()
```
